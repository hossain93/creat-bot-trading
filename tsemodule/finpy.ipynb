```python
from persiantools.jdatetime import JalaliDate
import datetime
import time
import os.path
def get_all_ri():
    #stock list
    df_stock_list = fpy.Build_Market_StockList(bourse = True, farabourse = False, payeh = False,
    detailed_list = True, show_progress = True,save_excel = False, save_csv = False)
    bourse_stocks = df_stock_list.index.tolist()

    #get all data
    j=0
    ri_history = pd.DataFrame({'stock': bourse_stocks, 'data': bourse_stocks, 'last_update':JalaliDate.today().isoformat()})
    ri_history = ri_history.set_index('stock')
    for i in bourse_stocks:
        print(j,end='-')
        df = fpy.Get_RI_History(stock = i,\
        start_date = (JalaliDate.today() - datetime.timedelta(10)).isoformat(),\
        end_date=JalaliDate.today().isoformat(),\
        ignore_date = False, show_weekday = True, double_date = False)
        ri_history.at[i,'data']=df
        j+=1
    return ri_history


start_time=time.perf_counter()
if os.path.isfile('ri_history.pkl'):
    status=input("اگر نیاز به آپدیت دارید مقدار y را وارد کنید در غیر این صورت n")
    if status=='y':
        ri_history=get_all_ri()
        ri_history.to_pickle("ri_history.pkl")
    if status=='n':
        ri_history = pd.read_pickle("ri_history.pkl")
else:
    ri_history=get_all_ri()
    ri_history.to_pickle("ri_history.pkl")
ri_history
print (time.perf_counter() - start_time, "seconds")
```

```python
from persiantools.jdatetime import JalaliDate
import datetime
import time
import os.path
def get_all_data():
    #stock list
    df_stock_list = fpy.Build_Market_StockList(bourse = True, farabourse = False, payeh = False,
    detailed_list = True, show_progress = True,save_excel = False, save_csv = False)
    bourse_stocks = df_stock_list.index.tolist()

    #get all data
    j=0
    all_data_price = pd.DataFrame({'stock': bourse_stocks, 'data': bourse_stocks, 'last_update':JalaliDate.today().isoformat()})
    all_data_price = all_data_price.set_index('stock')
    for i in bourse_stocks:
        print(j,end='-')
        df = fpy.Get_Price_History(stock = i,\
        ignore_date = True, adjust_price = True, show_weekday = True, double_date = True)
        all_data_price.at[i,'data']=df
        j+=1
    return all_data_price


start_time=time.perf_counter()    
if os.path.isfile('all_data_price.pkl'):
    status=input("اگر نیاز به آپدیت دارید مقدار y را وارد کنید در غیر این صورت n")
    if status=='y':
        all_data_price=get_all_data()
        all_data_price.to_pickle("all_data_price.pkl")
    if status=='n':
        all_data_price = pd.read_pickle("all_data_price.pkl")
else:
    all_data_price=get_all_data()
    all_data_price.to_pickle("all_data_price.pkl")
all_data_price
print (time.perf_counter() - start_time, "seconds")
```

# Market watch

```python
from persiantools.jdatetime import JalaliDate
import datetime
import time
import os.path
import finpy_tse as fpy
import pandas as pd

def get_MarketWatch(bourse = True,farabourse= False, payeh = False):
    df_stock_list = fpy.Build_Market_StockList(bourse = bourse, farabourse = farabourse, payeh = payeh,
    detailed_list = True, show_progress = True,
    save_excel = True, save_csv = False)
    bourse_stocks = df_stock_list.index.tolist()
    
    df_market_watch, df_order_book = fpy.Get_MarketWatch(save_excel = False)
    df=pd.DataFrame(columns=list(df_market_watch.columns),index=bourse_stocks)
    j=0
    forbiden_stock=[]
    for i in bourse_stocks:
        print(j,end='-')
        try:
            df.loc[i,:]=df_market_watch.loc[i,:]
            j+=1
        except Exception as err:
            j+=1
            forbiden_stock.append(i)
    forbiden_stock=pd.DataFrame(forbiden_stock)
    MarketWatch=df.dropna(axis=0, how='all')
    df.to_pickle("MarketWatch.pkl")
    return df ,forbiden_stock, df_order_book
MarketWatch,forbiden_stock,order_book=get_MarketWatch()
print('\n ',forbiden_stock)
MarketWatch
order_book
```
