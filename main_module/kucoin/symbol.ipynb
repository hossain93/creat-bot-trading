# margin

```python
import requests
import pandas as pd
url = "https://api.kucoin.com/api/v1/symbols"

payload={}
files={}
headers = {}

margin = requests.request("GET", url, headers=headers, data=payload, files=files)
margin=margin.json()
margin=margin['data']
#margin=margin[0]
# margin=pd.DataFrame([margin.keys()]).T
# margin = margin.rename({0: 'margin'}, axis=1)

margin=pd.DataFrame(margin)
pd.set_option('display.max_row', margin.shape[0]+1)
margin.drop(margin.loc[margin['isMarginEnabled']!=True].index, inplace=True)
margin.drop(margin.loc[margin['quoteCurrency']!="USDT"].index, inplace=True)
margin.reset_index(drop=True,inplace=True)
margin.rename(columns={"symbol":"pair"},inplace=True)
margin=margin['pair']
print(margin)
```

# future

```python
import requests
import pandas as pd
url = " https://api-futures.kucoin.com/api/v1/contracts/active"

payload={}
files={}
headers = {}

margin = requests.request("GET", url, headers=headers, data=payload, files=files)
margin=margin.json()
margin=margin['data']

margin=pd.DataFrame(margin)
pd.set_option('display.max_row', margin.shape[0]+1)
pd.set_option('display.max_column', margin.shape[0]+1)
margin.drop(margin.loc[margin['quoteCurrency']!="USDT"].index, inplace=True)
margin.rename(columns={"indexSymbol":"pair"},inplace=True)
margin=margin['pair']
margin
```
