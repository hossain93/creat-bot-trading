```python
import pandas as pd
import time
import datetime
import requests
def historicnobitex(symbol,timeframe,fromdate,todate):
        symbol     =  symbol
        tf         =  {'1h':'60','3h':'180','6h':'360','12h':'720','1d':'D','2d':'2D','3d':'3D'}
        duration   =  tf[timeframe]
        fromdate   =  int(time.mktime(datetime.datetime.strptime(fromdate, "%d/%m/%Y").timetuple()))
        todate     =  int(time.mktime(datetime.datetime.strptime(todate, "%d/%m/%Y").timetuple()))
        url = "https://api.nobitex.ir/market/udf/history?symbol=%s&resolution=%s&from=%s&to=%s" % (symbol,duration,fromdate,todate)

        payload={}
        headers = {}

        response = requests.request("GET", url, headers=headers, data=payload)

        df=response.json()

        t=df['t']
        df=pd.DataFrame(df)
        df.columns = ['s','dateTime','close','open','high','low', 'volume']
        df['dateTime'] = pd.to_datetime(df.dateTime, unit='s')
        df['date'] = df.dateTime.dt.strftime("%d/%m/%Y")
        df['time-utc'] = df.dateTime.dt.strftime("%H:%M:%S")
        df.drop(['s','dateTime'], inplace=True, axis=1)

        column_names = ["date", "time-utc", "open", "high", "low", "close", "volume"]
        df = df.reindex(columns=column_names)
        
        return df


symbol    =  'BTCIRT'
timeframe =  '1h'
fromdate  =  '19/03/2022'
todate    =  '25/03/2022'



df = historicnobitex(symbol,timeframe,fromdate,todate)
df=df.astype({'open': 'int64', 'high': 'int64', 'low': 'int64', 'close': 'int64', 'volume': 'float'})
pd.set_option('display.max_row', df.shape[0]+1)
#df[{'open','high'}]=df[{'open','high'}].astype('int64')
df
# t=df['t']
# df=pd.DataFrame(df)
# df.columns = ['s','dateTime','close','open','high','low', 'volume']
# df['dateTime'] = pd.to_datetime(df.dateTime, unit='s')
# df['date'] = df.dateTime.dt.strftime("%d/%m/%Y")
# df['time-utc'] = df.dateTime.dt.strftime("%H:%M:%S")
# df.drop(['s','dateTime'], inplace=True, axis=1)

# column_names = ["date", "time-utc", "open", "high", "low", "close", "volume"]
# df = df.reindex(columns=column_names)
```
